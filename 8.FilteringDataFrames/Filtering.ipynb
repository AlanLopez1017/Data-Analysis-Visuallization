```python
import pandas as pd
```

```python
houses = pd.read_csv("../DataAnalysis/data/kc_house_data.csv")
titanic = pd.read_csv("../DataAnalysis/data/titanic.csv")
netflix = pd.read_csv("../DataAnalysis/data/netflix_titles.csv", sep = "|", index_col = 0)
```

# Filtering dataframes with a boolean series

```python
df = titanic.head()
df.sex == "female"
```

```python
df[df.sex == "female"]
```

```python
df[df.survived == 1] # df[df["survived"] == 1]
```

```python
bools = [False, False, True, True, True]
df[bools]
```

# Filtering with comparison Operators

```python
titanic[titanic.survived == 1]
```

```python
titanic[titanic.age == "18"]
```

```python
titanic[titanic.pclass != 1].pclass.value_counts().plot(kind="bar")
```

```python
houses[houses["price"] > 5000000]
```

```python
houses[houses["bedrooms"] >= 10]
```

```python
houses[houses["sqft_living"] < 500]
```

# Between Method

```python
houses[houses["bedrooms"].between(5,7)]
```

```python
houses[houses["grade"].between(11,13)]
```

# isin() Method

```python
countries = ["India", "Japan", "South Korea"]
netflix[netflix["country"].isin(countries)].country.value_counts()
```

```python
mature = netflix["rating"].isin(["TV-MA", "R", "PG-13"])
```

```python
netflix[mature]
```

# Combining Conditions Using AND

```python
women = titanic["sex"] == "female"
died = titanic["survived"] == 0

titanic[women & died]
```

```python
houses[(houses["waterfront"] == 1) & (houses["price"] < 500000)]
```

```python
good_view = houses["view"] == 4
high_quality = houses["grade"] >= 11
smaller = houses["sqft_living"] <= 3000
houses[good_view & high_quality & smaller]
```
